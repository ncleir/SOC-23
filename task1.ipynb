# Task

A simple artificial neural network is composed of neurons which are conceptually derived from biological neurons. Each artificial neuron has inputs and produces a single output which can be sent to multiple other neurons. The mathematical function which relates the output of a neuron to it's input in pytorch's neural network module is given by $f(x)= w^{T}x + b$, where $w$ is the weight, $b$ is the bias and $x$ is the input
